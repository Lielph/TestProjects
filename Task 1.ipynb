# 1. Перенесите в столбцы D и E имена и отчества сотрудников из столбца C, в столбце C оставьте только фамилии.

```python
df.name = df.surname.str.split().str.get(1)
df.s_name = df.surname.str.split().str.get(2)
df.surname = df.surname.str.split().str.get(0)
```

```python
df.name.fillna("РОЗА",inplace=True)
df.s_name.fillna("МАРКОВНА",inplace=True)
df
```

```python
df["Пол"] = df["Пол"].astype("object")
df.info()

def gender(name):
    if str(name[-1]) == "А":
        return "женский"
    else:
        return "мужской"
df["Пол"] = df.s_name.apply(gender)
```

# 3. Укажите в столбце H возраст на дату приема (количество полных лет).

```python
df["Возраст на дату приема"] = df["Дата приема"].dt.year - df["Дата рождения"].dt.year
df
```

# 4. Укажите в столбце I, достиг ли сотрудник пенсионного возраста (мужчины с 60 лет, женщины с 55 лет) на дату приема, в форме да/нет.
---

```python
def pension_age(frame):
    if (frame["Пол"] == "женский" and frame["Возраст на дату приема"] >=55) or \
    (frame["Пол"] == "мужской" and frame["Возраст на дату приема"] >=60):
        return "да"
    else:
        return "нет"
df["Пенсионного возраста"] = df.apply(pension_age,axis=1)
df
```

```python
df.fillna(" ",inplace=True)
df
```

# 5. Заполните столбец L, если в столбце K дата увольнения пустая, то считайте стаж на дату отчета (ячейка F13).						
---

```python
df['Unix время'] = datetime.datetime.now().replace(hour=0, minute=0, second=0, microsecond=0)
def experience(frame):
    if frame["Дата увольнения"] == " ":
        return (frame['Unix время'] - frame["Дата приема"])
    else:
        return " "
df["Стаж"] = df.apply(experience, axis=1)
df["Стаж"] = df["Стаж"].astype("str")
df["Стаж"] = df["Стаж"].str.split().str.get(0)
df
```

```python
df["Стаж"].fillna(0,inplace=True)
```

```python
def tst(row):
    year, month, day = 0, 0 ,0
    if row != 0:
        year = int(row) // 365.25
        month = (int(row) % 365.25 // 30)
        return str(f"{int(year)}лет,{int(month)} месяцев")
    else:
        return " "
df["Стаж"] = df["Стаж"].apply(tst)
```
